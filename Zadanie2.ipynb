```python
import rospy
import math
from turtlesim.msg import Pose
```

```python
def initNode():
    rospy.init_node('pose_listener', anonymous=True)

def callback(data):
    rospy.loginfo("x = %s, y = %s, theta = %s", data.x, data.y, math.degrees(data.theta))
    
def print_pose():
    rospy.Subscriber("/turtle1/pose", Pose, callback)
    rospy.spin()
    
def unregisterSub():
    rospy.unregister("/turtle1/pose")
    
```

```python
initNode()
```

```python
print_pose()
```
